```python
class ScaleChords:
      def __init__(self, roots,thirds,fifths):
            self.roots=roots
            self.thirds=thirds
            self.fifths=fifths

def makeChords(r, scale):
    expandedScale = [r]
    sum = r;
    for x in scale:
        sum += x
        expandedScale.append(sum)
    roots = []
    thirds = []
    fifths = []
    for x in range(0, 7):
        root = expandedScale[x % 12];
        third = expandedScale[(x + 2) % 7];
        fifth = expandedScale[(x + 4) % 7];
        roots.append(root)
        thirds.append(third)
        fifths.append(fifth)
    return ScaleChords(roots,thirds,fifths)

def makeCircle(r):
    roots = []
    thirds =[]
    fifths =[]
    majorScale = [2,2,1,2,2,2]
    for x in range(0, 12):
        chords = makeChords(x, majorScale)
        roots = roots + chords.roots
        thirds = thirds + chords.thirds
        fifths = fifths + chords.fifths
    return ScaleChords(roots,thirds,fifths)

majorScale = [2,2,1,2,2,2]
CMajor = makeChords(0, majorScale)
FMajor = makeChords(5, majorScale)
GMajor = makeChords(7, majorScale)

Circle = makeCircle(0)

print(Circle.roots)
```
